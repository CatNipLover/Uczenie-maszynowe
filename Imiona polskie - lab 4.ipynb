```python
b = {}
for imie in imiona:
  string = ['.'] + list(imie) + ['.']
  for zn1, zn2 in zip(string, string[1:]):
    #print(zn1, zn2,   b.get((zn1, zn2), 0))
    b.get((zn1, zn2), 0)
    bigram = (zn1, zn2)
    b[bigram] = b.get((zn1, zn2), 0) + 1
b.get(('m', 'a'))
```

```python
b = {}
N_tab = torch.zeros((N, N), dtype=torch.int32)
for imie in imiona:
  string = ['.'] + list(imie) + ['.']
  for zn1, zn2 in zip(string, string[1:]):
    #print(zn1, zn2,   b.get((zn1, zn2), 0))
    b.get((zn1, zn2), 0)
    bigram = (zn1, zn2)
    b[bigram] = b.get((zn1, zn2), 0) + 1
    ix1 = stoi[zn1] - 1
    ix2 = stoi[zn2] - 1
    N_tab[ix1, ix2] +=1
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 20))
plt.imshow(N_tab, cmap ='Blues')

for i in range(0, N):
  for j in range(0, N):
    znaki = itos[i+1] + itos[j+1]
    plt.text(j, i, znaki, va='top', ha='center', color='green')
    plt.text(j, i, N_tab[i, j].item(), va='bottom', ha='center', color='green')
```

```python
random_seed = 173142
gen = torch.Generator().manual_seed(random_seed)
ix  = torch.multinomial(prob[0, :], num_samples = 1, replacement = True, generator = gen)
#print(ix.item())
znak = itos[ix.item()+1]
nick = '.' + znak
while (znak != '.') or (len(nick)<3):
  ix2 = torch.multinomial(prob[ix.item(), :], num_samples = 1, replacement = True, generator = gen)
  znak = itos[ix2.item() + 1]
  if(znak!='.'):
    ix = ix2
    nick = nick + znak
    if(len(znak)>= 14):
      break
print(nick.strip('.').capitalize())
```
