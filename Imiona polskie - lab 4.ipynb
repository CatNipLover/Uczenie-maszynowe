```python
imiona = open('imiona_polskie.txt', 'r').read().splitlines()
print(*imiona[:10])
print(f"Ile imiona w bazie:  {len(imiona)}")
print(f"najkrótsz imie: {min(len(w) for w in imiona)}")
imiona=list(filter(None, imiona))
print(f"Najdłuższe imie: {max(len(w) for w in imiona)}")
```

```python
imiona.append('Bądzisława')
imiona.append('Bądzisław')
imiona = [imie.lower() for imie in imiona]

print(*imiona[:10])
```

```python
znaki = sorted(list(set(''.join(imiona))))
print(*znaki)
znaki.insert(0, '.')
print(*sorted(znaki))

N = len(znaki)
print(f"liczba znaków: {N}")
```

```python
b = {}
N_tab = torch.zeros((N, N), dtype=torch.int32)
for imie in imiona:
  string = ['.'] + list(imie) + ['.']
  for zn1, zn2 in zip(string, string[1:]):
    #print(zn1, zn2,   b.get((zn1, zn2), 0))
    b.get((zn1, zn2), 0)
    bigram = (zn1, zn2)
    b[bigram] = b.get((zn1, zn2), 0) + 1
    ix1 = stoi[zn1] - 1
    ix2 = stoi[zn2] - 1
    N_tab[ix1, ix2] +=1
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 20))
plt.imshow(N_tab, cmap ='Blues')

for i in range(0, N):
  for j in range(0, N):
    znaki = itos[i+1] + itos[j+1]
    plt.text(j, i, znaki, va='top', ha='center', color='green')
    plt.text(j, i, N_tab[i, j].item(), va='bottom', ha='center', color='green')
```
